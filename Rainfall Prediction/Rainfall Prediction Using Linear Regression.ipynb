```python
import pandas as pd
import numpy as np
import sklearn as sk
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
data= pd.read_csv('austin_weather_final.csv')
```

```python
x= data.drop(['PrecipitationSumInches'],axis=1)
y= data['PrecipitationSumInches']
```

# Reshaping

```python
Y= y.values.reshape(-1,1)
```

```python
Y
```

```python
day_index = 798
days = [i for i in range(Y.size)]
```

```python
clf = LinearRegression()
clf.fit(x,Y)
```

```python
print('The precipitaion trend graph')
plt.scatter(days,Y,color= 'g')
plt.scatter(days[day_index],Y[day_index], color ='r')
plt.title('precipitaion level')
plt.xlabel('days')
plt.ylabel('precipitantin inches')
plt.show()
x_vis = x.filter(['TempAvgF','DewPointAvgF','HumidityAvgPercent','SealevelPressureAvgInches','VisibilityAvgMiles','WindAvgPH'])
```

```python
print('the precipitation vs attributes trend graph')
for i in range (x_vis.columns.size):
    plt.subplot(3,2,i+1)
    plt.scatter(days,x_vis[x_vis.columns.values[i][:100]],color='g')
    plt.scatter(days[day_index],x_vis[x_vis.columns.values[i]][day_index],color= 'r')
    plt.title(x_vis.columns.values[i])
plt.show()    
    
```
